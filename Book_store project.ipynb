```python
# 10. Remove comma from values in Price column and change the type to float64
print("10. Removing comma from Price and changing datatype to Float:") 
Book_store['Price'] = Book_store['Price'].str.replace(',', '').astype('float64')
Book_store.dtypes
```

```python
# 11. Remove comma from values in Total Ratings column and change the type to int64
print("11. Removing comma from Total Ratings and changing datatype to int:")
Book_store['Total Ratings'] = Book_store['Total Ratings'].str.replace(',', '').astype('int64')
Book_store.dtypes
```

```python
# 14. Replace missing values in Language column with the most occurring language
print("14. Replacing missing values in Launguage with most occuring language:")
most_common_language = Book_store['Language'].mode()[0]
Book_store['Language'].fillna(most_common_language, inplace=True)
Book_store.isnull().sum()
```

```python
# 15. Replace missing values in Pages column with median and convert data type to int64
print("15. Replacing missing values in Pages column with median and data type to int:")
median_pages = Book_store['Pages'].median()
Book_store.loc[Book_store['Pages'].isnull(), 'Pages'] = median_pages
Book_store['Pages'] = Book_store['Pages'].astype('int64')
Book_store.isnull().sum()
```

```python
# 43. Remove Outliers (using IQR method)
#Extra (above two questions with Outliners removed)
def remove_outliers(df, column, multiplier=1.5):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile) of the column
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Define the lower and upper bounds for outliers
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR

    # Remove outliers by filtering the column within the bounds
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df_cleaned

original_rows = len(Book_store)
Book_store = remove_outliers(Book_store, 'Price')
Book_store = remove_outliers(Book_store, 'Pages')
rows_removed = original_rows - len(Book_store)
print(f"43. Number of rows removed: {rows_removed}")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.boxplot(y=Book_store['Price'])
plt.title('Boxplot of Price (Outliers Removed)')
plt.subplot(1, 2, 2)
sns.boxplot(y=Book_store['Pages'])
plt.title('Boxplot of Pages (Outliers Removed)')
plt.tight_layout()
plt.show()
```
